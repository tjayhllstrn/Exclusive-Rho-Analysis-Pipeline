```python
c_Mxvsz = ROOT.TCanvas()
ROOT.gStyle.SetPalette(ROOT.kThermometer)
zbin_edgs = [0.66, 0.74]#, 0.79, 0.82, 0.85, 0.88, 0.91, 0.93, 0.96, 1]
nbins = len(zbin_edgs)-1
baseColor = 51
histograms = []
for i in range(nbins):
    chain1.Draw(f"Mx_true>>h{i}(100,0.5,3.5)",f"truepho2_pid==22&&truepho1_pid==22&&trueelectron_pid==11&&truepip_pid==211&&MCphoparent_samepi0==1&&MCtrue_containsNeutron==1&&Mdiphoton<0.16 && 0.115<Mdiphoton&& 1.2<Mh&&Mh<1.6 && {zbin_edgs[i]}<z&&z<{zbin_edgs[i+1]}","goff")

    hist = ROOT.gDirectory.Get(f"h{i}")
    hist.SetLineWidth(2)
    #color = baseColor + int((float(i) / (nbins - 1)) * 5)
    #hist.SetLineColor(color)
    histograms.append(hist)
c_Mxvsz.cd()
leg = ROOT.TLegend(0.6,0.2,0.9,0.85)
for i, hist in enumerate(histograms):
    leg.AddEntry(hist,f"{zbin_edgs[i]} < z < {zbin_edgs[i+1]}","l")
    if i ==0:
        hist.SetStats(0)
        hist.SetTitle("MC signal per z bin")
        hist.Draw("PLC")
    else:
        hist.Draw("PLC SAME")

leg.Draw()
c_Mxvsz.Draw()
```
