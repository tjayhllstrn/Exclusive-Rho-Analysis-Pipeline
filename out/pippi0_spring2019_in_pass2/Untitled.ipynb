```python
import ROOT
```

```python
filename = "nSidis_006616.root"
chain = ROOT.TChain("pippi0")
chain.Add(filename)
```

```python
[b.GetName() for b in chain.GetListOfBranches()]
```

```python
c = ROOT.TCanvas()

chain.Draw("t_z>>h(100,-1,1)","Mdiphoton<0.16 && 0.115<Mdiphoton && 0.85<Mx&&Mx<1.15","")
chain.Draw("t_elec>>h1(100,-1,1)","Mdiphoton<0.16 && 0.115<Mdiphoton && 0.85<Mx&&Mx<1.15","")
h1 = ROOT.gDirectory.Get("h1")
h1.SetLineColor(ROOT.kRed)
h = ROOT.gDirectory.Get("h")
h.SetLineColor(ROOT.kBlue)

leg = ROOT.TLegend(0.6,0.6,0.85,0.85)
leg.AddEntry(h,"t from z","l")
leg.AddEntry(h1,"t Greg","l")

#h.Draw()
h1.Draw("SAME")
leg.Draw()
c.Draw()
```
