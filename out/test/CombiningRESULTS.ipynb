```python
import ROOT
```

```python
f = ROOT.TFile("RESULTS_zbinning_0100.root")
f1 = ROOT.TFile("../BSA_Plots/pippi0_RGA_inbending_pass2_merged/RESULTS_zbinning_0100.root")
f.ls()
f1.ls()
```

```python
ROOT.gStyle.SetPalette(ROOT.kBird) #kColorPrintableOnGrey
gr0 = f.Get("gr_0")
gr1 = f.Get("gr_1")
gr2 = f.Get("gr_2")
gr3 = f1.Get("Graph")
#gr4 = f.Get("gr_4")
#gr5 = f.Get("gr_5")
graphs = {}
graphs["gr_0"] = gr0
graphs["gr_1"] = gr1
graphs["gr_2"] = gr2
graphs["gr_3"] = gr3
#graphs["gr_4"] = gr4
#graphs["gr_5"] = gr5
leg = f.Get("TPave")
c1 = f.Get("c1")
```

```python
len(graphs)
```

```python
ncolors = ROOT.TColor.GetNumberOfColors()
palette = ROOT.TColor.GetPalette()
colors = [int(palette.At(i)) for i in range(ncolors)]

c = ROOT.TCanvas("c","c",1000,800)
c.Divide(2,2)
c_idx = 1
for idx in range(len(graphs)):
    if idx ==0 or idx ==2 or idx==4:
        c.cd(c_idx)
        graphs[f"gr_{idx}"].GetYaxis().SetRangeUser(-1,1)
        graphs[f"gr_{idx}"].Draw("AP")
        c_idx += 1
    else:
        graphs[f"gr_{idx}"].Draw("P SAME")
c.cd(4)
leg.SetX1(0.1)
leg.SetY1(0.1)
leg.SetY2(0.9)
leg.SetX2(0.9)
leg.Draw()
c.Draw()
```

```python
canvas = ROOT.TCanvas()
canvas.SetTickx()
canvas.SetTicky()
canvas.SetGridx()
canvas.SetGridy()
for i in range(len(graphs)):
    if i ==0:
        graphs[f"gr_{i}"].GetYaxis().SetRangeUser(-0.2,0.5)
        graphs[f"gr_{i}"].Draw("AP")

    else:
        graphs[f"gr_{i}"].Draw("P SAME")

leg.SetX1(0.54)
leg.SetY1(0.7)
leg.SetY2(0.85)
leg.SetX2(0.88)
leg.SetBorderSize(1)
leg.AddEntry(graphs["gr_3"],"A_{sig} Chi2 Fit to M_{h} all cos#theta","p")
leg.Draw()
canvas.Draw()
```
