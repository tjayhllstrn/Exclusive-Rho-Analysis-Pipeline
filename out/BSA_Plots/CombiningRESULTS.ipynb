## No Cth binning

```python
def jitter_graph_x(graph, dx):
    x_vals = [graph.GetX()[i] for i in range(graph.GetN())]
    y_vals = [graph.GetY()[i] for i in range(graph.GetN())]
    for i in range(len(x_vals)):
        graph.SetPoint(i,x_vals[i]+dx,y_vals[i])
```

```python
nocth_fnames = ["pippi0_RGA_inbending_pass2_merged/tBinning_cthDep/RESULTS_t_elecbinning_1111.root"]

nocth_f1 = ROOT.TFile(nocth_fnames[0],"READ")
#nocth_f2 = ROOT.TFile(nocth_fnames[1],"READ")
#nocth_f3 = ROOT.TFile(nocth_fnames[2],"READ")
#nocth_f4 = ROOT.TFile(nocth_fnames[3],"READ")
nocth_f1.ls()
```

```python
gr_MLM_Mh = nocth_f1.Get("gr_0")
gr_chi2_Mh = nocth_f1.Get("gr_1")
gr_MLM_Mx = nocth_f1.Get("gr_2")
gr_chi2_Mx = nocth_f1.Get("gr_3")
dx = 0.003
jitter_graph_x(gr_MLM_Mx,-2*dx)
jitter_graph_x(gr_chi2_Mx,-dx)
jitter_graph_x(gr_MLM_Mh,0)
jitter_graph_x(gr_chi2_Mh,+dx)
```

```python
canvas = ROOT.TCanvas()
ROOT.gPad.SetGrid()

gr_MLM_Mx.SetLineColor(ROOT.kRed-6)
gr_MLM_Mx.SetMarkerColor(ROOT.kRed-6)
gr_chi2_Mx.SetLineColor(ROOT.kOrange-3)
gr_chi2_Mx.SetMarkerColor(ROOT.kOrange-3)
gr_MLM_Mh.SetLineColor(ROOT.kCyan-6)
gr_MLM_Mh.SetMarkerColor(ROOT.kCyan-6)
gr_chi2_Mh.SetLineColor(ROOT.kBlue-7)
gr_chi2_Mh.SetMarkerColor(ROOT.kBlue-7)

gr_MLM_Mx.GetYaxis().SetRangeUser(-0.1,0.5)
gr_MLM_Mx.GetXaxis().SetTitle("-t")
gr_MLM_Mh.GetYaxis().SetRangeUser(-0.1,0.5)
gr_MLM_Mx.SetTitle("RGA inbending #pi^{+}#pi^{0}")

leg = ROOT.TLegend(0.15,0.6,0.48,0.85)
#leg.AddEntry(gr_MLM_Mx,"A_{sig} MLM M_{x} Fit","l")
#leg.AddEntry(gr_chi2_Mx,"A_{sig} chi2 M_{x} Fit","l")
leg.AddEntry(gr_MLM_Mh,"A_{sig} MLM M_{h} Fit","l")
leg.AddEntry(gr_chi2_Mh,"A_{sig} chi2 M_{h} Fit","l")
leg.SetBorderSize(0)

#gr_MLM_Mx.Draw("AP")
#gr_chi2_Mx.Draw("P SAME")
gr_MLM_Mh.Draw("AP")
gr_chi2_Mh.Draw("P SAME")
leg.Draw()
canvas.Draw()
```
